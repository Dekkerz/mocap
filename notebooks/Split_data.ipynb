```python
#file has no header

X_path = os.path.join('smoking_data','smoking_input.csv')
y_path = os.path.join('smoking_data','smoking_targets.csv')

print(X_path,y_path)

X_data=pd.read_csv(X_path, header=None)
y_data=pd.read_csv(y_path, header=None)
```
